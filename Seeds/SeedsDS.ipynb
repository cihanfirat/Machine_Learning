```python
import pandas as pd
```

```python
#importing dataset and fill features
df = pd.read_csv("seeds_dataset.txt",delim_whitespace=True,
                names=['Area', 'Perimeter', 'Compactness', 'Length', 'Width', 'Asymmetry', 'Groove', 'Class'])
```

```python
df.head()
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
df.columns
```

PREPROCESSING

```python
X = df[['Area', 'Perimeter', 'Compactness', 'Length', 'Width', 'Asymmetry',
       'Groove']] #features
Y = df['Class'] #label
```

```python
set(Y) #Kama, Rosa and Canadian
```

```python
Y
```

SPLITTING

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2) #%20 for testing
```

MODELLING

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,Y_train)
```

```python
pred = dtc.predict(X_test)
```

```python
pred #predictions
```

```python
Y_test #actual
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,pred)*100)
```

Logistic Regression

```python
import pandas as pd
df = pd.read_csv("seeds_dataset.txt",delim_whitespace=True,
                names=['Area', 'Perimeter', 'Compactness', 'Length', 'Width', 'Asymmetry', 'Groove', 'Class'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
X=df[['Area', 'Perimeter', 'Compactness', 'Length', 'Width', 'Asymmetry',
       'Groove']]
Y = df["Class"]
```

```python
set(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
pred = lr.predict(X_test)
pred
```

```python
Y_test
```

```python
print(accuracy_score(Y_test,pred)*100)
```
